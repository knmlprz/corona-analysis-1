# Scrapowanie danych z koronawirusunas.pl

```python
import json
import urllib.request
import re
from bs4 import BeautifulSoup
```

```python
# load webpage
web = urllib.request.urlopen("https://www.koronawirusunas.pl")
soup = BeautifulSoup(web.read(), "html")

# Get all strings like 'var * = *;'
script = soup.find_all('script')[10]
js_vars = re.findall(r'var.*?=\s*(.*?);', script.string, re.DOTALL | re.MULTILINE)

# Parse each var to json
for js_var in js_vars:
    # Susbst key: val to "key" val
    r = re.sub(r'([\{\s,])(\w+)(:)', r'\1"\2"\3', js_var)
    
    # Fix ',' before } ] )
    r = re.sub(",(?!.*,)","",r)
    r = re.sub(",[ \t\r\n]*}", "}", r)
    r = re.sub(",[ \t\r\n]*\]", "]", r)
    
    #print(r)
    try:
        o = json.loads(r)
    except JSONDecodeError as e:
        # Exception 
        print("TTutaj jest błąd ")
```
